## Players data

**COMMENT:** Generate players' biographic information with birthdate's year being chosen randomly from 2004 to 2008. Noticably, each `player_id` has associated `academy_id` and `agent_id` (FKs).

```python
# 3. Generate players data
player_data = []
player_id = 1
for country in list(scouting_agencies.keys()):
    for city in city_dict[country]:
        for team in youth_teams_city[city]:
            for i in range(20):
                first_name = random.choice(first_names[country])
                last_name = random.choice(last_names[country])
                start_year = 2004
                end_year = 2008
                year = random.randint(start_year, end_year)
                month = random.randint(1, 12)
                day = random.randint(1, 28)
                birthdate = f"{year}-{month:02d}-{day:02d}"
                nationality = countries_nat[country]
                lang = random.choice(country_languages[country])
                height = random.randint(160, 210)
                weight = random.randint(50, 100)
                foot = random.choice(['left', 'right'])
                academy_id = random.choice(academies[academies["city"]==city]['academy_id'].tolist())
                agent_id = random.choice(agents[agents["agency_name"].isin(scouting_agencies[country])]['agent_id'].tolist())

                player_data.append([player_id, first_name, last_name, birthdate, city, country, nationality, lang, height, weight, foot, academy_id, agent_id])
                player_id +=1
players = pd.DataFrame(player_data, columns=['player_id', 'first_name', 'last_name', 'birthdate', 'city', 'country', 'nationality', 'lang', 'height', 'weight', 'foot', 'academy_id', 'agent_id'])
players.head()
```

## Award data

**COMMENT:** Randomly generate awards associated with special players. As a result, FK is `player_id`.

```python
# create list of unique player ids from players dataframe
player_ids = list(players['player_id'].unique())

# create list of award names
awards = ['Assist King', 'Golden Boot', 'Best Goalkeeper', 'Best Defender', 'Best Midfielder', 'Best Overall Player']

# randomly assign award to a player id

award_data = []
award_id = 1
for year in list(range(2020,2023)):
    for award in awards:
        for league in list(leagues.keys()):
            player_id = np.random.choice(players[players["country"]==league]["player_id"].to_list())
            award_data.append({
                'award_id':award_id,
                'player_id': player_id,
                'award_name': str(year) + " " + leagues[league] +  " "  + award,
                'award_year': year,
            })
            award_id += 1

# create pandas dataframe from award data
awards_df = pd.DataFrame(award_data)
awards_df.head()
```
